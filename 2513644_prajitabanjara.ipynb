```python
def happiness_category(x):
    if x < 4:
        return "Low"
    elif x <= 6:
        return "Medium"
    else:
        return "High"

df["Happiness Category"] = df["score"].apply(happiness_category)
df.head()
```

```python
top10 = df.sort_values("score", ascending=False).head(10)

import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.bar(top10["Country name"], top10["score"], color=
        'green')
plt.xticks(rotation=90)
plt.title("Top 10 Happiest Countries")
plt.show()
```

```python
bottom10 = df.sort_values("score").head(10)

import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(bottom10["Country name"], bottom10["score"], marker='o')
plt.xticks(rotation=90)
plt.title("Top 10 Unhappiest Countries")
plt.show()
```

```python
south_asia_list = ["Afghanistan", "Bangladesh", "Bhutan", "India",
                   "Maldives", "Nepal", "Pakistan", "Sri Lanka"]

sa_df = df[df["Country name"].isin(south_asia_list)].copy()
sa_df
```

```python
sa_df.to_csv("SouthAsia.csv", index=False)
```

```python
sa_df["Composite Score"] = (
    0.40 * sa_df["Log GDP per capita"] +
    0.30 * sa_df["Social support"] +
    0.30 * sa_df["Healthy life expectancy"]
)

sa_ranked = sa_df.sort_values("Composite Score", ascending=False)
sa_ranked
```

```python
top5 = sa_ranked.head(5)

import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.barh(top5["Country name"], top5["Composite Score"])
plt.title("Top 5 South Asian Countries by Composite Score")
plt.show()
```

```python
Q1 = sa_df["score"].quantile(0.25)
Q3 = sa_df["score"].quantile(0.75)
IQR = Q3 - Q1

low = Q1 - 1.5 * IQR
high = Q3 + 1.5 * IQR

outliers = sa_df[(sa_df["score"] < low) | (sa_df["score"] > high)]
outliers
```

```python
plt.figure(figsize=(8,5))
plt.scatter(sa_df["Log GDP per capita"], sa_df["score"], label="Normal")
plt.scatter(outliers["Log GDP per capita"], outliers["score"], color='red', label="Outliers")

plt.xlabel("Log GDP per capita")
plt.ylabel("Score")
plt.title("South Asia Outliers: Log GDP per Capita vs Score")
plt.legend()
plt.show()
```

```python
corr_freedom = sa_df[["Freedom to make life choices", "score"]].corr().iloc[0,1]
corr_generosity = sa_df[["Generosity", "score"]].corr().iloc[0,1]

corr_freedom, corr_generosity
```
